```python
x = df.iloc[:, 0:5].values
y = df.iloc[:,6].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0 )
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train , y_train)
y_pred = model.predict(x_test)
y_pred  # predicted value
```

```python
y_test #actual value
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter( y_pred , y_test )
plt.xlabel('y predicted')
plt.ylabel(' y actual')
plt.show()
```

```python
from sklearn.metrics import r2_score # this gives the accuracy of the model
r2_score( y_test , y_pred)
```

```python
df1 = pd.DataFrame({'Actual data' : y_test , 'Predicted data' : y_pred }) # creating a dataframe from predicted and actual values of y
df1
```

```python
df1.plot(figsize = (14,6))
plt.subplot(1 ,2,1)
sns.scatterplot( x = 'Actual data' , y = 'Predicted data' , data = df1 , label = 'graph 1')
plt.title('scatter plot')

plt.subplot(1,2,2)
sns.regplot( x = 'Actual data' , y = 'Predicted data' , data = df1 ,  marker="+"  , label =' graph2')
plt.title('Regression plot')
plt.legend()
plt.show() ,
```
